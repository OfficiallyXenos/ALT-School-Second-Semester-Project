# MAR (Missing At Random)


*   **Size**: The missing values in the `size` column are classified as MAR because they occur only when the `area_type` is "Plot", indicating that no residential unit exists. This pattern shows the missingness is linked to another known variable rather than the `size` itself.  It would also make sense because the `bath` and `balcony` entries are also NaN

* **Bath**:  Rows with missing `bath` values also consistently lacked `balcony`, suggesting both fields are skipped together in low-quality or incomplete listings. Since these rows still included `size` and non-plot properties, the missingness is not random but likely depends on other known features, making both columns MAR.

* **Balcony**: The missing values in balcony depend on other known info in the dataset  like `size`, `bath`, and probably even `price`.

* **Society**: The missing values in the society column likely depend on factors not captured in this dataset such as local infrastructure, as they appear across various area types and locations. Since the missingness isnt random but also not directly caused by the society name itself, i believe it is best categorized as MAR

# Missing Completely At Random (MCAR)

* **Location**: The `location` column has only one missing value out of thousands of rows, with no clear pattern or dependency on other variables. This strongly suggests the missingness is due to a random data entry error.






# Data Aggregation and Grouping

```python
url = 'https://raw.githubusercontent.com/nytimes/covid-19-data/refs/heads/master/us-counties-2020.csv'
covid_data = pd.read_csv(url)
#finding the mean of the cases by county and rounding off to 2 decimal places
cases_by_county = covid_data.groupby('county')['cases'].mean().round(2)
#renaming the 'cases' column to 'avg_cases'
cases_by_county.rename('avg_cases', inplace= True)
# soring the data in descending order
cases_by_county.sort_values(ascending= False, inplace= True)
#getting the top 10 counties
cases_by_county.head(10)
```

```python
cases_by_county.tail(5)
```

## Summary of the Summarization

To analyze the spread of COVID-19 across the United States in 2020, the dataset was grouped by county, and the average number of reported cases per day was calculated for each. This provides a county-by-county overview of how severely the pandemic impacted different regions on average.


---


## Insights from the Results
NYC recorded the highest daily average of COVID-19 cases (approx 215,822), followed by Los Angeles (approx 175,886) and Miami-Dade (~111,610).

Other significantly impacted counties include Maricopa, Broward, Cook, Harris, Riverside and San Bernadino , all averaging over 40,000 cases.

On the lower end, counties like Daggett, Hinsdale, Tianian, Kalawao and Loving had daily averages lower than 5 cases.

# Time Series Analysis

```python

import seaborn as sns #importing seaborn for viz
import matplotlib.pyplot as plt # importing matplotlib for viz
# converting the date column to datetime format
covid_data['date'] = pd.to_datetime(covid_data['date'])
print(covid_data['date'].dtype)
# creating a short-form 'month' column
covid_data['month'] = covid_data['date'].dt.month_name().str.slice(0, 3)
# subsetting COVID-19 data for Arizona
Arizona_data = covid_data[covid_data['state'] == 'Arizona']
# presenting a visual representation of the number of cases per month
sns.relplot(x= 'month', y = 'cases', kind= 'line', data= Arizona_data)
# setting the title of the visualization
plt.title('COVID-19 Cases per Month In Arizona')
# showing the visualization
plt.show()
```
